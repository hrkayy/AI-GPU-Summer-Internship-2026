```python
import torch
import time
import psutil

from huggingface_hub import login

from transformers import (
    pipeline,
    AutoTokenizer,
    AutoModel
)

# --------------------------
# LOGIN
# --------------------------

login("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")

print("=" * 60)
print("HUGGING FACE AI WORKSTATION")
print("=" * 60)

device = "cuda" if torch.cuda.is_available() else "cpu"

print("\nDevice:", device)

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# --------------------------
# TOKENIZER
# --------------------------

print("\nTOKENIZATION")

text = """
Artificial Intelligence is transforming
education and cybersecurity.
"""

tokenizer = AutoTokenizer.from_pretrained(
    "bert-base-uncased"
)

tokens = tokenizer.tokenize(text)

print("Tokens:")
print(tokens)

print("Token Count:", len(tokens))

# --------------------------
# EMBEDDINGS
# --------------------------

print("\nEMBEDDINGS")

model = AutoModel.from_pretrained(
    "bert-base-uncased"
)

inputs = tokenizer(
    text,
    return_tensors="pt",
    truncation=True
)

with torch.no_grad():
    outputs = model(**inputs)

embedding = outputs.last_hidden_state

print("Embedding Shape:", embedding.shape)

# --------------------------
# TEXT GENERATION
# --------------------------

print("\nTEXT GENERATION")

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

result = generator(
    "Machine Learning is",
    max_new_tokens=50,
    do_sample=True
)

print(result[0]["generated_text"])

# --------------------------
# SENTIMENT ANALYSIS
# --------------------------

print("\nSENTIMENT ANALYSIS")

sentiment = pipeline(
    "sentiment-analysis"
)

reviews = [
    "This workshop is amazing.",
    "The session was boring."
]

for review in reviews:
    print(review)
    print(sentiment(review))

# --------------------------
# SUMMARIZATION
# --------------------------

print("\nSUMMARIZATION")

summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

article = """
Artificial Intelligence is a branch of computer science focused on creating systems capable of performing tasks that typically require human intelligence. These tasks include learning, reasoning, problem solving, perception, and language understanding.
"""

summary = summarizer(
    article,
    max_length=40,
    min_length=10,
    do_sample=False
)

print(summary[0]["summary_text"])

# --------------------------
# TRANSLATION
# --------------------------

print("\nTRANSLATION")

translator = pipeline(
    "translation_en_to_fr",
    model="t5-small"
)

translated = translator(
    "Artificial Intelligence is transforming education."
)

print(translated[0]["translation_text"])

# --------------------------
# BENCHMARK
# --------------------------

print("\nBENCHMARK")

prompt = "Explain Transformers in AI."

start = time.time()

output = generator(
    prompt,
    max_new_tokens=100
)

end = time.time()

generated_text = output[0]["generated_text"]

tokens_generated = len(
    generated_text.split()
)

throughput = tokens_generated / (end - start)

print("Time:", round(end - start, 2), "sec")
print("Tokens/sec:", round(throughput, 2))

# --------------------------
# MEMORY
# --------------------------

print("\nMEMORY")

process = psutil.Process()

memory = process.memory_info().rss / 1024**2

print("RAM Usage:", round(memory, 2), "MB")

# --------------------------
# GPU MEMORY
# --------------------------

if torch.cuda.is_available():

    gpu_memory = (
        torch.cuda.memory_allocated()
        / 1024**3
    )

    print(
        "GPU Memory:",
        round(gpu_memory, 2),
        "GB"
    )

print("\nExecution Complete")
```
