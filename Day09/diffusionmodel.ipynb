```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("cute-curious-gray-and-white-kitten-in-a-long-shot-photo.jpg")

# Check if image loaded
if img is None:
    print("Image not found!")
else:
    print("Shape:", img.shape)

    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display image in Jupyter
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.show()

    # Save a copy
    cv2.imwrite("copy.jpg", img)
    print("Image saved as copy.jpg")
```
