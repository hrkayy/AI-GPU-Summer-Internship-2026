```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Load Dataset
data = pd.read_excel("student_marks_10000 (1).xlsx")

# Features
X = data[["Study_Hours", "Attendance", "Assignment_Score"]]

# ====================================
# REGRESSION - Predict Final Marks
# ====================================

y_reg = data["Final_Marks"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)

reg_predictions = reg_model.predict(X_test_reg)

print("========== REGRESSION ==========")
print("R2 Score:", r2_score(y_test_reg, reg_predictions))

# Predict Final Marks for a new student
new_student_reg = pd.DataFrame({
    "Study_Hours": [6],
    "Attendance": [85],
    "Assignment_Score": [75]
})

predicted_marks = reg_model.predict(new_student_reg)

print("Predicted Final Marks:")
print(predicted_marks[0])

# ====================================
# CLASSIFICATION - Predict Pass/Fail
# ====================================

le = LabelEncoder()
data["Result_Encoded"] = le.fit_transform(data["Result"])

y_cls = data["Result_Encoded"]

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X, y_cls, test_size=0.2, random_state=42
)

cls_model = LogisticRegression(max_iter=1000)
cls_model.fit(X_train_cls, y_train_cls)

cls_predictions = cls_model.predict(X_test_cls)

print("\n========== CLASSIFICATION ==========")
print("Accuracy:", accuracy_score(y_test_cls, cls_predictions))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test_cls, cls_predictions))

# Predict Pass/Fail for a new student
new_student_cls = pd.DataFrame({
    "Study_Hours": [6],
    "Attendance": [85],
    "Assignment_Score": [75]
})

prediction = cls_model.predict(new_student_cls)

if prediction[0] == 1:
    print("\nPrediction: Pass")
else:
    print("\nPrediction: Fail")
```
