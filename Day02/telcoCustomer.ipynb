```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, accuracy_score

# Load Excel Dataset
data = pd.read_excel("Telco_customer_churn.xlsx")

# Convert Churn Label to numeric
le = LabelEncoder()
data["Churn"] = le.fit_transform(data["Churn Label"])

# Convert Senior Citizen to numeric if needed
if data["Senior Citizen"].dtype == object:
    data["Senior Citizen"] = data["Senior Citizen"].map({
        "Yes": 1,
        "No": 0
    })

# -------------------------
# REGRESSION
# Predict Monthly Charges
# -------------------------

X_reg = data[["Tenure Months", "Senior Citizen"]]
y_reg = data["Monthly Charges"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)

reg_predictions = reg_model.predict(X_test_reg)

print("Regression R2 Score:")
print(r2_score(y_test_reg, reg_predictions))

new_customer_reg = pd.DataFrame({
    "Tenure Months": [24],
    "Senior Citizen": [0]
})

monthly_charge = reg_model.predict(new_customer_reg)

print("Predicted Monthly Charge:")
print(monthly_charge[0])

# -------------------------
# CLASSIFICATION
# Predict Churn
# -------------------------

X_cls = data[[
    "Tenure Months",
    "Monthly Charges",
    "Senior Citizen"
]]

y_cls = data["Churn"]

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=42
)

cls_model = LogisticRegression(max_iter=1000)
cls_model.fit(X_train_cls, y_train_cls)

churn_predictions = cls_model.predict(X_test_cls)

print("\nClassification Accuracy:")
print(accuracy_score(y_test_cls, churn_predictions))

new_customer_cls = pd.DataFrame({
    "Tenure Months": [24],
    "Monthly Charges": [75],
    "Senior Citizen": [0]
})

prediction = cls_model.predict(new_customer_cls)

if prediction[0] == 1:
    print("Customer Will Churn")
else:
    print("Customer Will Stay")
```
